##### Use shuffle

```python
number_options = [i for i in range(1,26)]

def generate_numbers_list():
    random.shuffle(number_options)
    final_list = number_options[:15]
    final_list.sort()
    
    print(final_list)
    
    return final_list        
    
```

```python
def main():
    file_name = './data/generated.csv'

    with open(file_name, mode='w', newline='') as csv_file:
        writer = csv.writer(csv_file)
        header_numbers = [f'bola{i+1}' for i in range(15)]
        header = ['concurso', *header_numbers]
        writer.writerow(header)

        for i in range(1, 3572):
            tmp = [i]
            numbers = generate_numbers_list()
            tmp = [i, *numbers]
            
            writer.writerow(tmp)
        
        print(f'Arquivo "{file_name}" foi criado com sucesso!')
```

```python
main()
```
